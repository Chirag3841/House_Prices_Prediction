```python
t= d2['Id']
```

```python
y = d1['SalePrice']
d1.drop(['SalePrice'], axis=1)
```

```python
c=pd.concat([d1,d2],axis=0,ignore_index=True)
```

```python
for i in c.columns:
     if c[i].dtype == "object":
        c[i] = c[i].fillna("None")
     
     else:
        if i == 'LotFrontage':
            c[i] = c.groupby('Neighborhood')[i].transform(lambda x: x.fillna(x.median()))
        elif i == 'GarageYrBlt':
            c[i] = c[i].fillna(c['YearBuilt'])
        else:
            c[i] = c[i].fillna(c[i].median())
c['TotalSF'] = c['TotalBsmtSF'] + c['1stFlrSF'] + c['2ndFlrSF'] 
c['TotalSF'] = c['TotalSF'].clip(upper=c['TotalSF'].quantile(0.99))
```

```python
c= pd.get_dummies(c,dummy_na=False)
```

```python
from sklearn.feature_selection import VarianceThreshold
columns = c.columns
selector = VarianceThreshold(threshold=0.01) 
c = selector.fit_transform(c)
c = pd.DataFrame(c, columns=columns[selector.get_support()])
```

```python
X_train = c.iloc[:len(y), :]
X_test = c.iloc[len(y):, :]
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
x=RandomForestRegressor(n_estimators=100,random_state=42)
x.fit(X_train,y)
```

```python
x1 = np.expm1(x.predict(X_test))
x1 = np.clip(x1, a_min=0, a_max=1e6)
```

```python
submission = pd.DataFrame({
    "Id": t,
    "SalePrice": x1
})

submission.to_csv("submission.csv", index=False)
```
